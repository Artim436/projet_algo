```python
## 1st case F0 = 7

Fe = 32
F0 = 7
a = 1/Fe
N = 16
T = 2*N*a

def f(x):
  return np.exp(2j*np.pi*F0*x)

def Sn(x):
  s = 0
  for n in range(-N, N):
    s += f(n*a)*np.exp(-2j*np.pi*n*a*x)
  return s


nb_point = 1000
X = np.linspace(0, N, nb_point)
Y = [abs(Sn(l)) for l in X]
K = [i for i in range(N)]
Xk = [i/T for i in range(N)]
Yk = [abs(Sn(i/T)) for i in range(N)]
plt.plot(X, Y, label=f"Sn(lambda) with nb_point = {nb_point}")
plt.plot(Xk, Yk, 'x', label=f"Sn(k/T) with nb_point = {N}")
plt.xlabel("lambda")
plt.legend()
plt.title("Sn(l) and Sn(k/T) with F0 = 7")
plt.show()

##2nd case f0 = 0.2

Fe = 32
F0 = 32*0.2
a = 1/Fe
N = 16
T = 2*N*a

def f(x):
  return np.exp(2j*np.pi*F0*x)

def Sn(x):
  s = 0
  for n in range(-N, N):
    s += f(n*a)*np.exp(-2j*np.pi*n*a*x)
  return s


nb_point = 1000
X = np.linspace(0, N, nb_point)
Y = [abs(Sn(l)) for l in X]
K = [i for i in range(N)]
Xk = [i/T for i in range(N)]
Yk = [abs(Sn(i/T)) for i in range(N)]
plt.plot(X, Y, label=f"Sn(lambda) with nb_point = {nb_point}")
plt.plot(Xk, Yk, 'x', label=f"Sn(k/T) with nb_point = {N}")
plt.xlabel("lambda")
plt.legend()
plt.title("Sn(l) and Sn(k/T) with f0 = 0.2")
plt.show()
```

##2 On the use of DFT for filtering

$\textbf{Exercice 2}$ $\textit{Circular convolution, computation of the inverse filter}$



$\textbf{1.}$

We have: 
$g = d * f = \sum_{n=0}^3\dfrac{1}{4}\delta_{na} *  \sum_{k \in \mathbb{Z}}f_k\delta_{ka}$

By using the formula from the course, we get:  
$g = \sum_{k \in \mathbb{Z}}f_k\sum_{n=0}^3\dfrac{1}{4}\delta_{(k+n)a} $
By posing the change of variable $m=k+n$

$g = \sum_{m \in \mathbb{Z}}\sum_{n=0}^3\dfrac{1}{4}f_{m-n}\delta_{ma}$

So we can identify 

$g_n = \sum_{k=0}^3\dfrac{1}{4}f_{n-k}$

$g$ is periodic with period $N$ because it is the case for $f$

$\textbf{2. 3.}$

We have that $\hat g = \hat d \hat f$

So $g = F^{-1}(\hat d \hat f)$

```python
a = 1/20

def dn(x):
    if x<=3*a:
        return 1/4
    return 0

def f(x):
    return np.sin(2*np.pi*x)

def g_dft(d, f):
    g_hat = []
    for d_hat, f_hat in zip(np.fft.fft(d), np.fft.fft(f)):
        g_hat.append(d_hat*f_hat)
    return g_hat


x = np.linspace(0, 2, 40)
d = [dn(xi) for xi in x]
f = [f(xi) for xi in x]
g_hat = g_dft(d[:-1], f[:-1])
g = np.fft.ifft(g_hat)

g_module = [gi.real for gi in g]
plt.plot(np.linspace(0, 2, 78), 2*g_module, label="g module")
plt.plot(np.linspace(0, 2, 78), 2*f[:-1], label="sin")
plt.xlabel("x")
plt.legend()
plt.title("Computation of g by the algorithm")
```

We get not the exact same curve because of $\delta_n$. We get $g_n$ insted of $f_n$.

$\textbf{4. 5.}$

First we calculate the fourier transform of $f$ thanks to the product of the fourier transform of the inverse filter and the fourier transform of g.

Then we just have to apply the inverse fourier transform to get $f$

```python
def f_hat(D, G, eps=1):
  res = []
  for d, g in zip(np.fft.fft(D), np.fft.fft(G)):
      if abs(d) > eps:
          res.append(g/d)
      else:
          res.append(g/eps)
  return np.array(res)

fh = f_hat(d[:-1], g)
fp = np.fft.ifft(fh)
fpr = [f.real for f in fp]
plt.plot(np.linspace(0, 1, 39), f[:-1], label="theorical")
plt.plot(np.linspace(0, 1, 39), fpr, label="pratical")
plt.title("Recover of f with eps = 1")
plt.legend()

```

$\textbf{Exercice 3}$ $\textit{Implementation of a finite impulse response filter}$

$\textbf{1.}$

We have, for $n \in \mathbb{Z}$

$h_n = \int_{-1/2}^{1/2}H(\lambda)\exp(2i\pi n \lambda)d\lambda $

$h_n = \int_{-1/4}^{1/4}\exp(2i\pi n \lambda)d\lambda$

So $h_0 = 1/2$ and for $n \neq 0$, $\dfrac{\exp(i\pi n/2)-\exp(-i\pi n/2)}{2i\pi n} = \dfrac{\sin(\pi n/2)}{\pi n}$

$\textbf{2.}$

```python
def h(n):
    if n ==0:
        return 1/2
    elif n%4 ==1 :
        return 1/(np.pi*n)
    elif n%4 == 3 :
        return -1/(np.pi*n)
    else :
        return 0

for n in range(-7, 8):
  print(f"h{n} = ", h(n))
```

```python
N = 7

def h(n):
    if n ==0:
        return 1/2
    elif n%4 ==1 :
        return 1/(np.pi*n)
    elif n%4 == 3 :
        return -1/(np.pi*n)
    else :
        return 0

def hc(n):
    add = -(N-1)/2
    return h(n+add)

def Hn(x):
    s = 0
    for n in range(-(N-1)//2,(N+1)//2):
        s += h(n)*np.exp(-2j*np.pi*n*x)
    return s

def Hnc(x):
    s = 0
    for n in range(N):
        s += hc(n)*np.exp(-2j*np.pi*n*x)
    return s


def H_theo(x):
    if -1/4<=x and 1/4>x:
        return 1
    else : 
        return 0


X = np.linspace(-1/2, 1/2, 100)



Y_theo = np.array([H_theo(x) for x in X])
Y_Hn = np.array([Hn(x) for x in X])
Y_Hnc = np.array([Hnc(x) for x in X])



plt.plot(X, np.angle(Y_Hn), label = "non causal")
plt.plot(X, np.angle(Y_Hnc), label = "causal")
plt.legend()
plt.title("arguments")
plt.show()


plt.plot(X, abs(Y_Hn), label="modulus with causal filter")
plt.plot(X, Y_theo, label="H")
plt.title("theorical and pratical functions")
plt.legend()
plt.show()
plt.plot(X, abs(Y_Hnc), label="modulus with non causal filter")
plt.plot(X, Y_theo, label="H")
plt.title("theorical and pratical functions")
plt.legend()
plt.show()
```

The translation affects the phase but not the modulus of the function.

We can see that the phase is piecewise linear in case of causal filter because the fourier transform of the signal is real.
In the other case the signal is piecewise constant.

$\tilde{h_n} = \int_{-1/4}^{1/4}\exp(i\pi (2n-1) \lambda)d\lambda$


So for all $n$, $\tilde{h_n} = \dfrac{\exp(i\pi (2n-1)/4)-\exp(-i\pi (2n-1)/4)}{i\pi (2n-1)} = \dfrac{\sin(\pi(2n-1)/4)2}{\pi (2n-1)}$

```python
def h_tilde(n):
  return 2*np.sin(np.pi*(2*n-1)/4)/(np.pi*(2*n-1))

for n in range(-2, 4):
  print(f"h_tilde_{n} = ", h_tilde(n))
```
